```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
samples_data = {
    'L_la': ['L_la_1', 'L_la_2', 'L_la_3', 'L_la_4'],
    'E_la': ['E_la_1', 'E_la_2', 'E_la_3', 'E_la_4'],
    'BL_la': ['BL_la_1', 'BL_la_2', 'BL_la_3', 'BL_la_4'],
    'L_bs': ['L_bs_1', 'L_bs_2', 'L_bs_3', 'L_bs_4'],
    'E_bs': ['E_bs_1', 'E_bs_2', 'E_bs_3', 'E_bs_4'],
    'BL_bs': ['BL_bs_1', 'BL_bs_2', 'BL_bs_3', 'BL_bs_4'],
    'L_ao': ['L_ao_1', 'L_ao_2', 'L_ao_3', 'L_ao_4'],
    'E_ao': ['E_ao_1', 'E_ao_2', 'E_ao_3', 'E_ao_4'],
    'BL_ao': ['BL_ao_1', 'BL_ao_2', 'BL_ao_3', 'BL_ao_4']
}

def combine_samples(data_frame):
    lst_dataframes = list()
    conditions = list()
    for key in samples_data:
        cols = samples_data[key]
        df_condition = pd.DataFrame(data_frame, columns=cols)
        sum_df_condition = df_condition.median(axis=1)
        sum_df_condition = pd.DataFrame(sum_df_condition, columns=[key])
        lst_dataframes.append(sum_df_condition)
        conditions.append(key)
    concat_df = pd.concat(lst_dataframes, axis=1)
    return concat_df, conditions

def prepare_data(file_path):
    norm_genes = pd.read_csv(file_path, sep="\t")
    y_labels = norm_genes["Gene"].tolist()
    norm_genes = norm_genes.drop(columns=['Gene'])
    norm_genes, x_labels = combine_samples(norm_genes)
    return norm_genes, y_labels, x_labels
```

```python
# metrics: cityblock, cosine
def plot_cluster_genes(dataframe, z_score, file_name, fig_title, y_labels, x_labels):
    sns.set(font_scale=2.25)
    fig_size = (25,60)
    g = sns.clustermap(dataframe,
                   row_cluster=True,
                   col_cluster=True,
                   cmap="RdBu_r",
                   xticklabels=x_labels,
                   yticklabels=y_labels,
                   figsize=fig_size,
                   metric='cosine',
                   z_score=z_score,
                  )
    g.fig.suptitle(fig_title)
    g.ax_heatmap.set_xlabel("Samples")
    g.ax_heatmap.set_ylabel("Genes")

    for a in g.ax_col_dendrogram.collections:
        a.set_linewidth(3)

    for a in g.ax_row_dendrogram.collections:
        a.set_linewidth(3)
    g.savefig('{}.pdf'.format(file_name), dpi=300, figsize=fig_size)
    g.savefig('{}.png'.format(file_name), dpi=300, figsize=fig_size)
```

```python
path = "Normalized_counts_all_DEgenes_for_clustering.tabular"
normalised_counts_all_genes, y_labels, x_labels = prepare_data(path)
plot_cluster_genes(normalised_counts_all_genes, 0, "median_heatmap_with_Seaborn_clustermap_python", 'Median of all samples: DEgenes for clustering_n=131', y_labels, x_labels)
```
